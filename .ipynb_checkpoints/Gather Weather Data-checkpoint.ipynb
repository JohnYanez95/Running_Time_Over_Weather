# Libraries

```python
# Web Scraping
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# Date Times
from datetime import datetime,timedelta

# Data Frames
import pandas as pd

# Text Formating
import re

# SQL
from sqlalchemy import create_engine

# Graphics
import matplotlib.pyplot as plt

# Training/Testing Sets
from sklearn.model_selection import train_test_split

# Regression
from sklearn.linear_model import LinearRegression
```

# Loading in Running_Times

```python
Running_Times = pd.read_csv('Running Times.txt')
```

Formatting date_time to a datetime data type and reformatting duration to a float:

```python
# Applying date_time type to date_time column of Running Times
Running_Times['date_time'] = Running_Times['date_time'].apply(pd.to_datetime)

# Function used to change our duration format from minutes:seconds to a float of minutes
def duration_to_float(row):
    hr,mi = map(int,row.split(':'))
    return hr+mi/60
Running_Times['duration'] = Running_Times['duration'].apply(duration_to_float)
Running_Times
```

# Web Scraping

The following function is very specific to our web scraping. The information we desire comes from 

        'https://www.wunderground.com/history/daily/us/tx/austin/KAUS/date/'+str(dt.year)+\
        '-'+str(dt.month)+'-'+str(dt.day)'
where dt.year, dt.month, and dt.day are varied by the iteration over the 'Running Times' Data Set's column: 'date_time'. 

The information gathered from 'https://www.wunderground.com' has hourly time stamps that revolve around minute 53. So the solution to determining which hourly time stamp coincides with a current iteration of my 'Running Times' Data Set is the closest_time to our the 'midpoint date time', where the 'midpoint data time' is calculated by adding half of the 'duration time' to 'date time' from the 'Running Times' Data Set. 

$\underline{\text{Note}}:$ This calculation will be observed within the iteration of

        for dt,dur in zip(Running_Times['date_time'],Running_Times['duration'])
within the cell following the next.

```python
def to_closest_time(dt,minute=53):
    # Calculates the datetime of an hour back, the current hour, and hour ahead
    h1 = datetime(dt.year,dt.month,dt.day,dt.hour-1,minute)
    h2 = datetime(dt.year,dt.month,dt.day,dt.hour,minute)
    h3 = datetime(dt.year,dt.month,dt.day,dt.hour+1,minute)
    # Calculates the differences, and determines the least time difference to the minute mark
    v1 = abs(dt-h1)
    v2 = abs(dt-h2)
    v3 = abs(dt-h3)
    if v1 == min(v1,v2,v3): return h1
    elif v2 == min(v1,v2,v3): return h2
    else: return h3
```

Within the following cell I will iterat over the rows of the Running Times Data Set to Construct a Weather Data Set.

$\underline{\text{Note}}:$ If the index falls out of range when web scraping try restarting the cell a couple more times.


```python
bi = FirefoxBinary(r'C:\Program Files (x86)\Mozilla Firefox\\firefox.exe')
driver = webdriver.Firefox(firefox_binary=bi)
out_file = open('Weather Data.csv', 'w')
out_file.write('date_time,poll_time,temperature,dew_point,humidity,wind,wind_speed,wind_gust,pressure,precip,condition\n')
for dt,dur in zip(Running_Times['date_time'],Running_Times['duration']):
    # Pulling the Data from Online
    url = 'https://www.wunderground.com/history/daily/us/tx/austin/KAUS/date/'\
    +str(dt.year)+'-'+str(dt.month)+'-'+str(dt.day)
    driver.get(url)
    table = WebDriverWait(driver,20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "table")))
    table = BeautifulSoup(table[1].get_attribute('outerHTML'), 'lxml').find('table')
    
    
    # Determining Midpoint Date Time
    dur = dur/2
    mm,ss = int(dur),round(dur%1*60)
    midpoint_date_time = dt+timedelta(minutes=mm,seconds=ss)
    
    # Formating our Midpoint Date Time to match for a row within the html data
    val = to_closest_time(midpoint_date_time,53)
    if val.hour//12==1:
        val = str(val.hour%12)+':53 PM'
    else: val = str(val)+':53 AM'    

    
    # Takes the HTML rows and converts them to a comma seprated rows
    table_rows = table.find_all('tr')
    for row in table_rows:
        columns = row.find_all('td')
        if columns and columns[0].text.strip()==val:
            output_row = [str(dt)]
            for column in columns:
                output_row.append(column.text.strip())
            # Some cleaning and formatting of the text before writing:
            row = ','.join(output_row)
            row = re.sub('\s', '', row) + '\n'
            out_file.write(row)
out_file.close()
```

Due to the nature of my web scraping I must reformat the 'date_time' column to correspond to the date_time column of the Running Times data set. This will be accomplish by a string concatenation and applying the datetime data type to the rows of WeatherData.csv.

We also would like to strip the temperature, dew_point, humidity, wind_speed, wind_gust, pressure, and precip of their string values and reformat them as floats.

```python
Weather_Data = pd.read_csv('Weather Data.csv')
# Reformatting the 'date_time' column to correspond to the date_time column of the Running Times data set
f = lambda x: x[:-8]+' '+x[-8:]
Weather_Data['date_time'] = Weather_Data['date_time'].apply(f)
Weather_Data['date_time'] = Weather_Data['date_time'].apply(pd.to_datetime)
# Stripping unnecessary strings from data set
g = lambda row: int(re.match('\d+',row).group())
variables = ['temperature','dew_point','humidity','wind_speed','wind_gust','pressure','precip']
for var in variables:
    Weather_Data[var] = Weather_Data[var].apply(g)
Weather_Data
```

# To SQL
Establish a connection to our Local PostgreSQL data base:

```python
cnxn_string = ("postgresql+psycopg2://postgres:1596@localhost:5432/sqlda")
engine = create_engine(cnxn_string)
engine
```

Send our data sets to our SQL database.

$\underline{\text{Note}}:$ When sending data sets from python to PostgreSQL make sure that the tables and columns are labeled with no spaces and within all lower-case letters. Doing otherwise will lead to easily avoidable headaches and confusion when attempting to load the data using SQL.

```python
Weather_Data.to_sql('weatherdata',engine,index=False,if_exists='replace')
Running_Times.to_sql('runningtimes',engine,index=False,if_exists='replace')
```

# Inner Join of Running Times and Weather Data

```python
query = """
    SELECT WD.*,RT.DURATION 
    FROM RUNNINGTIMES RT
    INNER JOIN WEATHERDATA WD ON WD.DATE_TIME=RT.DATE_TIME
    ;
"""
join = pd.read_sql_query(query,engine)
join
```

# Regressions

This is where the project concludes, for now! A simple multiple linear regression was formed on 60\% of the data set using the weather data as predictors to the variable duration/running_time; no significant results were found yet. The most apparent reasons for the lack of results could be:
* the lack of data
* and the lack of variation within my data.

I wish to come back to this project once more runs are established, perhaps within 30 or 60 data points.

```python
predictors = ['temperature','dew_point','humidity','wind','wind_speed','wind_gust','pressure']
outcome = 'duration'
X = pd.get_dummies(join[predictors])
y = join[outcome]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.6,random_state=0)
LinReg = LinearRegression().fit(X_train,y_train.values)
print('{:<13} : {:<30.3f}\n'.format('Intercept',LinReg.intercept_))
print('{:<13}'.format('Coefficients'))
for name,coef in zip(X.columns,LinReg.coef_):
    print('{:<13} : {:<30.3f}'.format(name,coef))
```

The following cell shows the results of our regression:

```python
fig,ax = plt.subplots(figsize=(16,8))
plt.plot(join.date_time,join.duration,'b-',label='True Running Times')
plt.plot(join.date_time,LinReg.predict(X.values),'r--',label='Prediction Running Times')
plt.legend()
plt.show()
```
